```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
```

```python
# passing the file's path to a variable
baracuda_audio = "audio1/car_horn.wav"
```

```python
ipd.Audio(baracuda_audio)
```

```python
# file'loading
barac, sr = librosa.load(baracuda_audio)
```

```python
barac.shape
```

```python
# MFCCS extraction
mfccs = librosa.feature.mfcc(barac, n_mfcc=13, sr=sr)
```

```python
mfccs.shape
```

```python
# plotting
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs,
                        x_axis="time",
                        sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
# calculate delta and delta2 MFCCs
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
```

```python
delta_mfccs.shape
```

```python
# plotting
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs,
                        x_axis="time",
                        sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
# plotting
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs,
                        x_axis="time",
                        sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
# concatenatin delta_mfccs and delta2_mfccs
concatenate_mfccs = np.concatenate((mfccs, delta_mfccs, delta2_mfccs))
```
